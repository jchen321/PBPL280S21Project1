```python
import pandas as pd
import geopandas as gpd
```

```python
tract_income = gpd.read_parquet('ca_tract_incomes.parquet')
```

```python
tract_income
```

```python
sd_income = tract_income[tract_income.GEOID.str.startswith('06073')] 
```

```python
sb_income = tract_income[tract_income.GEOID.str.startswith('06071')] 
```

```python
riverside_income = tract_income[tract_income.GEOID.str.startswith('06065')]
```

```python
orange_income = tract_income[tract_income.GEOID.str.startswith('06059')]
```

```python
income_counties = pd.concat([sd_income,sb_income,riverside_income,orange_income], axis = 0) 
```
